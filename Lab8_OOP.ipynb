```python
# 1. Creating a Class and Object in Python

class Car:
    def __init__(self, brand, model, year):
        """Initializes the attributes of the car."""
        self.brand = brand
        self.model = model
        self.year = year

    def display_info(self):
        """Returns a formatted string of the car details."""
        return f"{self.year} {self.brand} {self.model}"

my_car = Car("Toyota", "Corolla", 2022)

print("--- Procedure 1: Car Class Output ---")
print(my_car.display_info())
print("\n" + "="*40 + "\n")



# TRY IT: Laptop Class Implementation

class Laptop:
    def __init__(self, brand, processor, ram):
        """Initializes the attributes for the Laptop class."""
        self.brand = brand
        self.processor = processor
        self.ram = ram

    def get_specs(self):
        """Returns the specifications in a clean, formatted string."""
        return (f"Laptop Specifications:\n"
                f"- Brand: {self.brand}\n"
                f"- Processor: {self.processor}\n"
                f"- RAM: {self.ram}GB")

my_laptop = Laptop("ASUS", "Intel Core i7", 16)

print("--- 'Try It' Task: Laptop Class Output ---")
print(my_laptop.get_specs())
```

```python
# 2. Implementing Encapsulation in Python

class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.__balance = balance 

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"New balance: ${self.__balance}"
        return "Invalid deposit amount."

    def get_balance(self):
        return f"Balance: ${self.__balance}"

account = BankAccount("Alice", 5000)

print("--- Procedure 2: BankAccount Output ---")
print(account.deposit(1500))
print(account.get_balance())
print("\n" + "="*40 + "\n")



# TRY IT: StudentRecord Implementation

class StudentRecord:
    def __init__(self):
        self.__grades = []

    def add_grade(self, grade):
        """Method to add a grade to the private list."""
        if 0 <= grade <= 100:
            self.__grades.append(grade)
            print(f"Added grade: {grade}")
        else:
            print(f"Error: {grade} is an invalid grade.")

    def get_all_grades(self):
        """Method to retrieve the list of all grades."""
        return self.__grades

my_record = StudentRecord()

my_record.add_grade(88)
my_record.add_grade(92)
my_record.add_grade(95)

print("\n--- 'Try It' Task: Student Record Output ---")
print(f"Full Grade List: {my_record.get_all_grades()}")
```

```python
# 3. Applying Inheritance in Python

class Person:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        return f"Hi, I'm {self.name}."

class Student(Person):
    def introduce(self):
        return f"Hi, I'm {self.name}, and I'm a student."

person = Person("Xandre")
student = Student("Dre")

print("--- Procedure 3: Person & Student Output ---")
print(person.introduce())
print(student.introduce())
print("\n" + "="*40 + "\n")


# TRY IT: Device and Smartphone Implementation

class Device:
    def __init__(self, device_name):
        self.device_name = device_name

    def power_on(self):
        """General power on message."""
        return f"The {self.device_name} is now powered on."

class Smartphone(Device):
    def power_on(self):
        """Overrides the base method with a specific message."""
        return f"The {self.device_name} smartphone is booting up... Welcome!"

generic_device = Device("Generic Tablet")
my_phone = Smartphone("iPhone 15")

print("--- 'Try It' Task: Inheritance Output ---")
print(generic_device.power_on()) 
print(my_phone.power_on())       
```

```python
# 4. Utilizing Polymorphism in Python

class Shape:
    def area(self):
        return "Calculating area..."

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

circle = Circle(5)
rectangle = Rectangle(4, 6)

print("--- Procedure 4: Shape Polymorphism Output ---")
print(f"Circle area: {circle.area()}")
print(f"Rectangle area: {rectangle.area()}")
print("\n" + "="*40 + "\n")


# TRY IT: Device Polymorphism Implementation

class Device:
    def power_on(self):
        return "Device is powering on..."

class Smartphone(Device):
    def power_on(self):
        return "Smartphone: System loading... Mobile network connected."

class Laptop(Device):
    def power_on(self):
        return "Laptop: BIOS initializing... Desktop environment ready."

def start_device(device):
    """
    This function demonstrates polymorphism. It accepts any object 
    derived from Device and invokes its power_on() method.
    """
    print(device.power_on())

print("--- 'Try It' Task: Device Polymorphism Output ---")

my_phone = Smartphone()
my_laptop = Laptop()

start_device(my_phone)
start_device(my_laptop)
```

```python
# 6. Working with Class and Static Methods

class MathOperations:
    @staticmethod
    def square(num):
        """A static method that returns the square of a number."""
        return num * num

try:
    print("--- Procedure 6: MathOperations Output ---")
    user_input = int(input("Enter a number to square: "))
    result = MathOperations.square(user_input)
    print(f"Square of {user_input}: {result}")
except ValueError:
    print("Error: Please enter a valid integer.")

print("\n" + "="*40 + "\n")


# TRY IT: TemperatureConverter Implementation

class TemperatureConverter:
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        """
        Static method to convert Celsius to Fahrenheit.
        Formula: (Celsius * 9/5) + 32
        """
        return (celsius * 9/5) + 32

try:
    print("--- 'Try It' Task: Temperature Converter Output ---")
    c_input = float(input("Enter a temperature in Celsius: "))

    f_result = TemperatureConverter.celsius_to_fahrenheit(c_input)
    
    print(f"{c_input}°C is equal to {f_result:.2f}°F")
    
except ValueError:
    print("Error: Invalid input. Please enter a numerical temperature.")
```

```python
# 7. Operator Overloading in Python

class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def __add__(self, other):
        """Overloads the + operator to combine balances."""
        return BankAccount(self.balance + other.balance)

account1 = BankAccount(5000)
account2 = BankAccount(3000)
total_balance = account1 + account2  

print("--- Procedure 7: BankAccount Overloading ---")
print(f"Total Balance: P{total_balance.balance}")
print("\n" + "="*40 + "\n")


# TRY IT: Vector Class with + and - Overloading

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        """Overloads the + operator for Vector addition."""
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        """Overloads the - operator for Vector subtraction."""
        return Vector(self.x - other.x, self.y - other.y)

    def __str__(self):
        """Helper method to print the vector nicely."""
        return f"Vector({self.x}, {self.y})"


v1 = Vector(10, 5)
v2 = Vector(3, 8)


v_sum = v1 + v2  
v_diff = v1 - v2 

print("--- 'Try It' Task: Vector Overloading Output ---")
print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}")
print(f"Addition (v1 + v2): {v_sum}")
print(f"Subtraction (v1 - v2): {v_diff}")
```

```python
# 8. Abstract Classes and Methods in Python

from abc import ABC, abstractmethod

class Appliance(ABC):
    @abstractmethod
    def turn_on(self):
        """Abstract method that must be implemented by subclasses."""
        pass

class WashingMachine(Appliance):
    def turn_on(self):
        return "Washing machine is now running."

wm = WashingMachine()

print("--- Procedure 8: Abstract Class Output ---")
print(wm.turn_on())
print("\n" + "="*40 + "\n")



# TRY IT: Refrigerator Subclass Implementation

class Refrigerator(Appliance):
    def turn_on(self):
        """Implementing the abstract method specifically for Refrigerator."""
        return "Refrigerator is now cooling."

my_fridge = Refrigerator()

print("--- 'Try It' Task: Abstract Class Extension Output ---")
print(my_fridge.turn_on())
```

```python
# 9. Implementing Method Resolution Order (MRO)

class A:
    def show(self):
        return "Class A"

class B(A):
    def show(self):
        return "Class B"

class C(A):
    def show(self):
        return "Class C"

class D(B, C):
    """
    Class D inherits from B first, then C. 
    MRO will prioritize B over C.
    """
    pass 

obj = D()

print("--- Procedure 9: MRO Behavior ---")
print(f"Method result for obj.show(): {obj.show()}")
print("\n" + "="*40 + "\n")


# TRY IT: Exploring MRO with Multiple Classes

class Alpha:
    def identify(self):
        return "Alpha"

class Beta:
    def identify(self):
        return "Beta"

class Gamma:
    def identify(self):
        return "Gamma"

class Omega(Alpha, Beta, Gamma):
    """
    Omega inherits from Alpha, Beta, and Gamma in that specific order.
    """
    def identify(self):
        return "Omega"

print("--- 'Try It' Task: MRO Exploration Output ---")
print(f"Calling Omega().identify(): {Omega().identify()}")

print("\nInternal Method Resolution Order (MRO) for class Omega:")
for index, cls in enumerate(Omega.mro()):
    print(f"{index + 1}. {cls}")
```
