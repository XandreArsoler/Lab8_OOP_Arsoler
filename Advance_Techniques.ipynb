```python
from abc import ABC, abstractmethod
from dataclasses import dataclass

# --- 1. Abstract Base Classes ---
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        return "Bark! Bark!"

# --- 2. Data Classes ---
@dataclass
class Student:
    name: str
    age: int
    course: str

# --- 3. Custom Exception Handling ---
class InvalidAgeError(Exception):
    pass

def set_age(age):
    if age < 0:
        raise InvalidAgeError("Age cannot be negative")
    return age

# --- 4. Property Decorators ---
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius

    @property
    def fahrenheit(self):
        return (self._celsius * 9/5) + 32

# --- 5. Composition Over Inheritance ---
class Engine:
    def start(self):
        return "Engine started"

class Car:
    def __init__(self):
        self.engine = Engine() # Composition: Car "has an" engine

    def drive(self):
        status = self.engine.start()
        return f"{status}... Car is moving"

# --- EXECUTION ---
print("--- Advanced Techniques Outputs ---")

# 1. ABC
dog = Dog()
print(f"ABC (Animal): {dog.sound()}")

# 2. Data Class
std = Student("Alice", 20, "CS")
print(f"Data Class: {std}")

# 3. Custom Exception
try:
    print(f"Age set to: {set_age(-5)}")
except InvalidAgeError as e:
    print(f"Custom Exception: {e}")

# 4. Property
temp = Temperature(25)
print(f"Property (Fahrenheit): {temp.fahrenheit}")

# 5. Composition
my_car = Car()
print(f"Composition: {my_car.drive()}")
```
