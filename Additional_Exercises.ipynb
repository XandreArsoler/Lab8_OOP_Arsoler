```python
class Book:
    def __init__(self, title, author):
        self.title = str(title)
        self.author = author
        self.is_available = True

class Library:
    def __init__(self):
        self.books = {}

    def add_book(self, title, author):
        self.books[title.lower()] = Book(title, author)
        return f"Book '{title}' added successfully."

    def borrow_book(self, title):
        title_key = title.lower()
        if title_key not in self.books:
            return "Error: Book does not exist."
        if not self.books[title_key].is_available:
            return "Error: Book is already borrowed."
        
        self.books[title_key].is_available = False
        return f"You have successfully borrowed '{self.books[title_key].title}'."

    def return_book(self, title):
        title_key = title.lower()
        if title_key in self.books:
            self.books[title_key].is_available = True
            return f"Returned '{self.books[title_key].title}'."
        return "Error: This book does not belong to this library."

    def display_status(self):
        print("\n--- Library Inventory ---")
        for book in self.books.values():
            status = "Available" if book.is_available else "Borrowed"
            print(f"Title: {book.title} | Status: {status}")

lib = Library()
lib.add_book("Python Basics", "John Smith")
lib.add_book("Clean Code", "Robert Martin")
print(lib.borrow_book("Python Basics"))
print(lib.borrow_book("Python Basics"))
lib.display_status()
```

```python
class Employee:
    def __init__(self, name, hourly_wage, hours_worked):
        if hourly_wage < 0 or hours_worked < 0:
            raise ValueError("Wage and hours cannot be negative.")
        self.name = name
        self.hourly_wage = hourly_wage
        self.hours_worked = hours_worked

    def calculate_salary(self):
        base_hours = min(160, self.hours_worked)
        overtime_hours = max(0, self.hours_worked - 160)
        
        base_pay = base_hours * self.hourly_wage
        overtime_pay = overtime_hours * (self.hourly_wage * 1.5)
        
        return base_pay, overtime_pay

employees = [
    Employee("James", 50, 170), 
    Employee("Kazel", 40, 150)    
]

for emp in employees:
    base, over = emp.calculate_salary()
    print(f"Name: {emp.name} | Base: ${base} | Overtime: ${over} | Total: ${base + over}")
```

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

class ShoppingCart:
    def __init__(self):
        self.items = {}
        self.total_price = 0

    def add_item(self, product, quantity):
        if quantity > product.stock:
            return f"Error: Only {product.stock} units of {product.name} left."
        
        self.items[product.name] = self.items.get(product.name, 0) + quantity
        product.stock -= quantity
        self.total_price += product.price * quantity
        return f"Added {quantity} x {product.name} to cart."

    def view_cart(self):
        print(f"Cart Contents: {self.items}")
        print(f"Total Price: ${self.total_price}")

laptop = Product("Laptop", 1000, 5)
cart = ShoppingCart()
print(cart.add_item(laptop, 2))
cart.view_cart()
```
